```python
#이름으로 데이터프레임 세개로 쪼개기
a=iris[iris['species'] == 'setosa']
b=iris[iris['species'] == 'versicolor']
c=iris[iris['species'] == 'virginica']
```

```python
#이상치 구하기
a1 =a.describe().T['25%'] - 1.5*(a.describe().T['75%']-a.describe().T['25%'])
a2 =a.describe().T['75%'] + 1.5*(a.describe().T['75%']-a.describe().T['25%'])
b1 =b.describe().T['25%'] - 1.5*(b.describe().T['75%']-b.describe().T['25%'])
b2 =b.describe().T['75%'] + 1.5*(b.describe().T['75%']-b.describe().T['25%'])
c1 =c.describe().T['25%'] - 1.5*(c.describe().T['75%']-c.describe().T['25%'])
c2 =c.describe().T['75%'] + 1.5*(c.describe().T['75%']-c.describe().T['25%'])
```

```python
# 이상치 필터링후 이름 다시 넣어주기
na=a[(a > a1) & (a < a2)]
nb=b[(b > b1) & (b < b2)]
nc=c[(c > c1) & (c < c2)]
na.species = 'setosa'
nb.species = 'versicolor'
nc.species = 'virginica'
```

```python
# 필터링한 데이터프레임 합치기
newiris = pd.concat([na,nb,nc])
```

```python
iris.describe()
```

```python
# 숫자가 안맞는걸 보니 필터링 확인
newiris.describe()
```

```python
# 답
newiris.groupby(newiris.species).agg(["mean","std"])
```

### 3-a

```python
mpg = sns.load_dataset('mpg')
```

```python
mpg
```

```python
# 열 추가하기
mpg['hp_per_cc'] = mpg.horsepower/mpg.displacement
```

### 3-b

```python
# 이름을 리스트화
mf = []
mn = []
mrn=[]
mfn = mpg.name.to_list()
```

```python
# 맨앞부분은 이름으로 사용하기 위해 다른 리스트에 추가, 나머지는 같은 방법으로 모델명에 추가
for i in mfn :
    tr = ''.join(i)
    ttr = tr.split()
    mf.append(ttr[0])
    mn.append(ttr[1:])
for i in mn:
    tr=' '.join(i)
    mrn.append(tr)
```

```python
mpg['manufacturer'] = mf
mpg['model'] = mrn
```

```python
del mpg['name']
```

```python
mpg
```

### 3-c

```python
mpg.groupby("cylinders")[["mpg"]].mean()
```

### 3-d

```python
mpg.groupby("origin")[["hp_per_cc"]].mean()
```

### 3-e

```python
# 모델이 5개이상인 제조사 찾기
bm = mpg.manufacturer.value_counts()[mpg.manufacturer.value_counts() > 5]
```

```python
# 인덱스만 필요하니 인덱스만 리스트화
brm = list(bm.index)
brm
```

```python
# 찾은 인덱스 + isin 활용
brrm = mpg[mpg.manufacturer.isin(brm)]
brrm
```

```python
answer = brrm.pivot_table("mpg","manufacturer")
```

```python
answer5 = answer.sort_values(by='mpg', ascending=False)
answer5
```
