```python
colors = ['red', 'green', 'pink', 'blue']
icons = ['leaf', 'fire', 'certificate', 'thumbs-up']
```

```python
mapping = folium.Map(location=[abc['위도'].mean(), abc['경도'].mean()],
                    zoom_start=12)
for i in abc.index:
    if abc['상호명'][i].find('스타벅스') != -1 or abc['상호명'][i].find('starbucks') != -1 :
        n_color = colors[0]
        n_icon = icons[0]
    elif abc['상호명'][i].find('빽다방') != -1 :
        n_color = colors[1]
        n_icon = icons[1]
    elif abc['상호명'][i].find('투썸') != -1  or abc['상호명'][i].find('TWOSOMECOFFEE') != -1 or abc['상호명'][i].find('Twosomeplace') != -1 or abc['상호명'][i].find('투썸플레이스') != -1 :
        n_color = colors[2]
        n_icon = icons[2]
    else : 
        n_color = colors[3]
        n_icon = icons[3]
    folium.Marker(
        location=[abc['위도'][i], abc['경도'][i]],
        popup=abc['상호명'][i],
        icon=folium.Icon(color = n_color, icon= n_icon)
    ).add_to(mapping)
mapping
```
