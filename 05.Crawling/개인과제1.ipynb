### 데이터프레임으로 바꾸고 합치기

```python
station_files = glob('./과제data/지역*.xls')
tmp_raw = []
for file_name in station_files:
    tmp = pd.read_excel(file_name, header=2)
    tmp_raw.append(tmp)

station_raw = pd.concat(tmp_raw)
```

```python
station_raw
```

### 휘발유 가격으로 정렬

```python
station_raw =station_raw.sort_values(by = '휘발유').reset_index(drop=True)
station_raw
```

```python
# import url
from urllib.parse import urlparse
import requests
# 카카오 키 불러오기
key_fd = open('kakaomapkey.txt', mode='r')
kmap_key = key_fd.read(100)
key_fd.close()
```

### 위도경도 표시하기

```python
lat_list =[] ; lng_list = [] 
for i in range(len(station_raw)):
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + station_raw['주소'][i]
    result = requests.get(urlparse(url).geturl(),headers={"Authorization": "KakaoAK "+kmap_key}).json()
    lat = float(result['documents'][0]['address']['y'])
    lng = float(result['documents'][0]['address']['x'])
    lat_list.append(lat)
    lng_list.append(lng)

station_raw['lat'] = lat_list
station_raw['lng'] = lng_list
```

### 지도에 표시하기

```python
colors = ['green', 'red','black']
```

### 가장 싼 주유소는 빨간색, 가장 비싼 주유소는 검은색으로 표시, 나머지는 녹색으로 표시

```python
import folium
map = folium.Map(location=[station_raw['lat'].mean(), station_raw['lng'].mean()], zoom_start=12.4)

for n in range(len(station_raw)) :
    if n == 0:
        k = 1
    elif n == 9:
        k = 2 
    else:
        k = 0
    folium.Marker([station_raw['lat'][n], 
    station_raw['lng'][n]],popup=station_raw['상호'][n],
    tooltip=f'{station_raw["휘발유"][n]}', icon=folium.Icon(color=colors[k])).add_to(map)

map
```

#### bs4 이용해서 크롤링 하기(selenium은 화면에 나오지 않는 정보를 긁어오지 못하는데 bs4는 긁어올 수 있다 )

```python
o_name = [] ; o_price =[]; o_addr =[] ; s = list(range(0,20000,2))
driver = webdriver.Chrome('./chromedriver')
driver.maximize_window()
driver.get('http://www.opinet.co.kr/user/main/mainView.do')
time.sleep(1)
driver.find_element_by_css_selector('.ic_m1').click()
time.sleep(2)
gu = '강서구'
dongs = ['등촌동','화곡동', '마곡동']
for dong in dongs:
    region = driver.find_element_by_id('SIGUNGU_NM0')
    region.send_keys(gu)
    time.sleep(1)
    s_dong = driver.find_element_by_id('DONG_NM')
    s_dong.send_keys(dong)
    time.sleep(1)
    driver.find_element_by_xpath('//*[@id="searRgSelect"]').click()
    time.sleep(1)
    for i in range(len(driver.find_elements_by_class_name('rlist'))):
        if len(driver.find_elements_by_class_name('rlist')[i].text) == 0 :
            break
        driver.find_elements_by_class_name('rlist')[i].find_element_by_tag_name('a').click()
        time.sleep(1)
        html = driver.page_source # 페이지의 elements모두 가져오기
        soup = BeautifulSoup(html, 'html.parser') 
        o_name.append(soup.find('div', 'header').get_text())
        o_addr.append(soup.find('label', {'id' : 'rd_addr'}).get_text())
        o_price.append(driver.find_elements_by_class_name('price')[s[i]].text)
        
```

```python
c_df = pd.DataFrame({
    '상호' : o_name,
    '가격' : o_price,
    '주소' : o_addr
})
```

```python
c_df = c_df.sort_values(by = '가격').reset_index(drop=True)
```
