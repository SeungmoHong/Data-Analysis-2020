# 데이터 시각화

### 시각화 패키지 맷플롯리브 소개

```python
# import
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

#### 라인플롯

```python
plt.title("Plot")
# 플롯 값에 하나만 주면 Y좌표로 인식 x는range(y)만큼
plt.plot([1, 4, 9, 16])
plt.show()
```

```python
# 플롯만 추가하면 줄 하나 더 생성 가능
plt.title("Plot")
plt.plot([1, 4, 9, 16])
plt.plot([1, 2, 8, 22])
plt.show()
```

```python
plt.title("Line Plot")
plt.grid(True) # 표같은 배경 생성
plt.plot([1, 4, 9, 16])
plt.show()
```

```python
plt.title("Line Plot")
plt.grid(True)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot([1, 4, 9, 16])
plt.show()
```

```python
plt.title("x ticks")
plt.grid(True)
# x좌표, y좌표 순
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.show()
```

```python
plt.title("x 틱스") # 그냥 사용하면 한글 깨짐
plt.grid(True)
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.show()
```

#### 한글 폰트 사용 방법

```python
import matplotlib.font_manager

matplotlib.font_manager._rebuild()
sorted([f.name for f in matplotlib.font_manager.fontManager.ttflist if f.name.startswith("Malgun")])
```

```python
import matplotlib.font_manager

matplotlib.font_manager._rebuild()
sorted([f.name for f in matplotlib.font_manager.fontManager.ttflist if f.name.startswith("Nanum")])
```

```python
# 폰트 설정
mpl.rc('font', family='Malgun Gothic')
# 유니코드에서  음수 부호설정
mpl.rc('axes', unicode_minus=False)
```

```python
## 한글폰트 적용 성공
plt.title('한글 제목')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.xlabel("엑스축 라벨")
plt.ylabel("와이축 라벨")
plt.show()
```
