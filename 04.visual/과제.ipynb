```python
plt.plot(g.index, g['일반의 수'], 'ro-',label="일반의")
plt.plot(g.index, g['인턴 수'], 'bo-',label="인턴")
plt.plot(g.index, g['레지던트 수'], 'yo-',label="레지던트")
plt.plot(g.index, g['전문의 수'], 'mo-',label="전문의")
plt.title('서울 및 광역시 종합병원 전문의 수', loc='center', pad=10, fontsize=16, color='g')
plt.xlabel('시')
plt.ylabel('의사 수')
plt.legend()
plt.show()
```

#### 추가문제

```python
import seaborn as sns
iris = sns.load_dataset('iris')
titanic = sns.load_dataset("titanic")
mpg = sns.load_dataset('mpg')
mf = []
mn = []
mrn=[]
mfn = mpg.name.to_list()
for i in mfn :
    tr = ''.join(i)
    ttr = tr.split()
    mf.append(ttr[0])
    mn.append(ttr[1:])
for i in mn:
    tr=' '.join(i)
    mrn.append(tr)
mpg['manufacturer'] = mf
mpg['model'] = mrn
del mpg['name']
```

```python
iris.boxplot(by='species')
plt.title("figure size : (11, 10)")
plt.tight_layout(pad=3, h_pad=1)
plt.title("각 Feature의 종 별 데이터에 대한 Box Plot")
plt.show()
```

```python
df = titanic.sex.value_counts()
df.plot.pie(autopct='%.2f%%')
plt.title("승객별 남녀 비율")
plt.axis('equal')
plt.show()
```

```python
df = titanic.embark_town.value_counts()
df.plot.pie(autopct='%.2f%%')
plt.title("승객별 승선지 비율")
plt.axis('equal')
plt.show()
```
