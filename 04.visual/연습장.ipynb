```python
# import
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
iris = sns.load_dataset('iris')
```

```python
iris = sns.load_dataset('iris')
ds1 = iris[iris.species == 'setosa']
ds2 = iris[iris.species == 'versicolor']
ds3 = iris[iris.species == 'virginica']
fig = plt.figure()
axes1 = fig.add_subplot(3, 1, 1); axes2 = fig.add_subplot(3,1,2)
axes3 = fig.add_subplot(3, 1, 3)
axes1.boxplot(ds1['sepal_length'], ds1['sepal_width'],ds1['petal_length'],ds1['petal_width'])
axes2.boxplot(ds2['sepal_length'], ds2['sepal_width'],ds2['petal_length'],ds2['petal_width'])
axes3.boxplot(ds3['sepal_length'], ds3['sepal_width'],ds3['petal_length'],ds3['petal_width'])
axes1.set_title('setosa'); axes2.set_title('versicolor')
axes3.set_title('virginica')
fig.tight_layout()
```

```python
x1 = iris[iris.species == 'setosa'].sepal_length
y1 = iris[iris.species == 'setosa'].sepal_width
z1 = iris[iris.species == 'setosa'].petal_length
v1 = iris[iris.species == 'setosa'].petal_width
plt.title('iris_setosa')
plt.boxplot([x1,y1,z1,v1], labels=['sepal_length','sepal_width','petal_length','petal_width'])
plt.show()
```

```python
x2 = iris[iris.species == 'versicolor'].sepal_length
y2 = iris[iris.species == 'versicolor'].sepal_width
z2 = iris[iris.species == 'versicolor'].petal_length
v2 = iris[iris.species == 'versicolor'].petal_width
plt.title('iris_versicolor')
plt.boxplot([x2,y2,z2,v2], labels=['sepal_length','sepal_width','petal_length','petal_width'])
plt.show()
```

```python
x3 = iris[iris.species == 'virginica'].sepal_length
y3 = iris[iris.species == 'virginica'].sepal_width
z3 = iris[iris.species == 'virginica'].petal_length
v3 = iris[iris.species == 'virginica'].petal_width
plt.title('iris_virginica')
plt.boxplot([x3,y3,z3,v3], labels=['sepal_length','sepal_width','petal_length','petal_width'])
plt.show()
```
